### Topic 2: Solution Sets of Linear Systems


Let's visualize the following equations from Section 1.4 Practice Problem 1 (page 47) to show if the planes intersect or not. The equations we are plotting are: 

$x_1 + 4x_2 - 5x_3 = 0 \\
 2x_1 - x_2 + 8x_3 = 9$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Now we can define the equations as functions. To do this we need to follow the python function conventions as follows: 
def plane1(x1, x2):
    return (5 * x1 - x2) / 4

def plane2(x1, x2):
    return (9 - 2 * x1 + 8 * x2) / 8
```

```python
# For plotting, we'll create a grid of points using np.linspace and np.meshgrid. 
# Starting with np.linspace, we can generate a vector with evenly spaced values
# from -10 to 10 with 50 samples. We can do this for both x1 and x2, and then 
# generate a meshgrid using these two vectors. 

x1 = np.linspace(-10, 10, 50)
x2 = np.linspace(-10, 10, 50)
X1, X2 = np.meshgrid(x1, x2)

# Now we can compute corresponding z values for each plane as follows using our
# previously defined functions:
X3_X1 = plane1(X1, X2)
X3_X2 = plane2(X1, X2)
```

```python
# Now for plotting, we will start by creating the figure and axes as follows:
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the first plane
ax.plot_surface(X1, X2, X3_X1, alpha=0.5, cmap='viridis')

# Plot the second plane
ax.plot_surface(X1, X2, X3_X2, alpha=0.5, cmap='inferno')

# Set labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.set_title('Planes')

# Adjust the viewing angle and perspective. The elev parameter sets the 
# elevation angle (vertical viewing angle), and the azim parameter sets the 
# azimuth angle (horizontal viewing angle). You can modify these values as per
# your preference to achieve the desired view of the crossing planes. 

ax.view_init(elev=30, azim=45)

# Show the plot
plt.show()
```

### Topic 1: Linear Independence

In the module, we learned how to determine if vectors are linearly independent using a row reduction technique. Now let's see how we can do that using python! Using the vectors from Section 1.7 Exercises 1-4, we can determine if the vectors are linearly independent. 

1. $\begin{bmatrix}
5 \\
0 \\
0 \\
\end{bmatrix},
\begin{bmatrix}
7 \\
2 \\
-6 \\
\end{bmatrix},
\begin{bmatrix}
9 \\
4 \\
-8 \\
\end{bmatrix}$

```python
import numpy as np

# Define the vectors from problem 1 as rows of a matrix
vectors = np.array([[5, 0, 0], [7, 2, -6], [9, 4, 8]])

# Perform row reduction manually
reduced_form = vectors.astype(float)
rows, cols = reduced_form.shape

for i in range(rows - 1):
    for j in range(i + 1, rows):
        factor = reduced_form[j, i] / reduced_form[i, i]
        reduced_form[j, :] -= factor * reduced_form[i, :]

# Check if any row is all zeros
if np.any(np.all(np.isclose(reduced_form, 0), axis=1)):
    print("The vectors are linearly dependent.")
else:
    print("The vectors are linearly independent.")
```

### Topic 2: Introduction to Linear Transformations

To further investigate linear transformations, we will look at an example with transforming an image. This is similar to Example #3 from Section 1.8 (Page 66). Here we'll load an image of a sheep (you can also change this to any image by providing the URL) and apply a transformation matrix that shears the image.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

# Load an image
image = io.imread('https://encrypted-tbn2.gstatic.com/licensed-image?q=tbn:ANd9GcQUzr3uUKYI_-stvBEzPcrjxWbPlYpKcdo5QKglbhv_yMc4eq7xRasKMe0Q-lFn6CuKUnzqEjUajYl3YGY')

# Set the amount of shear
shear_factor = 0.5

# Define the transformation matrix
shear_matrix = np.array([[1, shear_factor, 0],
                         [0, 1, 0],
                         [0, 0, 1]])

# Get the image dimensions
height, width, channels = image.shape

# Initialize the transformed image with zeros
transformed_image = np.zeros_like(image)

# Perform the affine transformation
for y in range(height):
    for x in range(width):
        # Apply the transformation matrix to the pixel coordinates
        transformed_coords = shear_matrix @ [x, y, 1]
        transformed_coords = transformed_coords[:2] / transformed_coords[2]
        
        # Round the transformed coordinates to integers
        transformed_x, transformed_y = np.round(transformed_coords).astype(int)
        
        # Check if the transformed coordinates are within the image boundaries
        if 0 <= transformed_x < width and 0 <= transformed_y < height:
            # Assign the pixel value from the original image to the transformed image
            transformed_image[transformed_y, transformed_x, :] = image[y, x, :]

# Display the original and transformed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(transformed_image)
axes[1].set_title('Transformed Image')
axes[1].axis('off')
plt.show()
```

## Day 4: Understanding Vector Spaces and Subspaces

### Topic 1: Vector Spaces

Let's start by reviewing the definition of a vector space from page 192. From the textbook, we learned that a vector space is a nonempty set $V$ of objects called vectors, on which are defined two operations, called addition and multiplication by scalars. The following must hold true for all vectors $\textbf{u}$, $\textbf{v}$, and $\textbf{w}$ in $V$ and for all scalars $c$ and $d$. 

1. The sum of $\textbf{u}$ and $\textbf{v}$, denoted by $\textbf{u} + \textbf{v}$ is in $V$.
2. $\textbf{u} + \textbf{v}$ = $\textbf{v} + \textbf{u}$ 
3. $(\textbf{u} + \textbf{v}) + \textbf{w}$ = $\textbf{u} + (\textbf{v} + \textbf{w})$
4.  There is a zero vector $\textbf{0}$ in $V$ such that $\textbf{u} + \textbf{0} = \textbf{u}$
5. For each $\textbf{u}$ in $V$, there is a vector $\textbf{-u}$ in $V$ such that $\textbf{u} + \textbf{-u} = \textbf{0}$
6. The scalar multiple of $\textbf{u}$ by $c$, denoted by $c\textbf{u}$, is in $V$. 
7. $c(\textbf{u} + \textbf{v}) = c\textbf{u} + c\textbf{v}$
8. $(c+d)\textbf{u} = c\textbf{u} + d\textbf{u}$
9. $c(d\textbf{u}) = (cd)\textbf{u}$
10. $1\textbf{u} = \textbf{u}$

Review the examples below and try different $\textbf{u}$, $\textbf{v}$, and $\textbf{w}$ vectors and different $c$ and $d$ scalar values. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors u and v
u = np.array([1, 3])
v = np.array([3, -3])
w = np.array([-1,-3])

# Define the origin
origin = np.array([0, 0])

# Property 1: The sum of u and v is in V
sum_uv = u + v

# output the sums
sum_uv_str = ', '.join(str(element) for element in sum_uv)
print(f"u + v = [{sum_uv_str}]")

# Plot Property 1
plt.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(*u, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(*origin, *sum_uv, angles='xy', scale_units='xy', scale=1, color='g', label='u + v')

plt.xlim(-2, 6)
plt.ylim(-2, 6)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

```python
# Property 2: u + v = v + u
sum_vu = v + u

# output the sums
sum_uv_str = ', '.join(str(element) for element in sum_uv)
print(f"u + v = [{sum_uv_str}]")

sum_vu_str = ', '.join(str(element) for element in sum_vu)
print(f"v + u = [{sum_vu_str}]")

# Plot Property 2
plt.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(*u, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(*origin, *sum_uv, angles='xy', scale_units='xy', scale=1, color='g', label='u + v')

plt.quiver(*sum_uv, *u, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(*u, *sum_vu, angles='xy', scale_units='xy', scale=1, color='m', label='v + u')

plt.xlim(-2, 6)
plt.ylim(-2, 6)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

```python
# Property 3: (u + v) + w = u + (v + w)
sum_vw = v + w
sum_uv_w = sum_uv + w
u_sum_vw = u + sum_vw

# output the sums
u_sum_vw_str = ', '.join(str(element) for element in u_sum_vw)
print(f"u + (v + w) = [{u_sum_vw_str}]")

sum_uv_w_str = ', '.join(str(element) for element in sum_uv_w)
print(f"(u + v) + w = [{sum_uv_w_str}]")

# Plot Property 3
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot data on the first subplot
ax1.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='r', label='u')
ax1.quiver(*u, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v')
ax1.quiver(*sum_uv, *w, angles='xy', scale_units='xy', scale=1, color='g', label='w')
ax1.quiver(*u, *sum_vw, angles='xy', scale_units='xy', scale=1, color='c', label='v + w')
ax1.quiver(*origin, *u_sum_vw, angles='xy', scale_units='xy', scale=1, color='m', label='u + (v + w)')
ax1.set_xlim(-4, 7)
ax1.set_ylim(-4, 7)
ax1.set_title('c(u + v)')
ax1.legend()

# Plot data on the second subplot
ax2.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='r', label='u')
ax2.quiver(*u, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v')
ax2.quiver(*sum_uv, *w, angles='xy', scale_units='xy', scale=1, color='g', label='w')
ax2.quiver(*origin, *sum_uv, angles='xy', scale_units='xy', scale=1, color='c', label='u + v')
ax2.quiver(*origin, *sum_uv_w, angles='xy', scale_units='xy', scale=1, color='m', label='(u + v) + w')
ax2.set_xlim(-4, 7)
ax2.set_ylim(-4, 7)
ax2.set_title('cu + cv')
ax2.legend()

# Adjust spacing between subplots
plt.tight_layout()

# Show the figure with subplots
plt.show()
```

### Topic 2: Rank

In Module 5 we covered how to determine the rank of a matrix. Here we'll see how we can compute the rank using the built-in numpy function:

 `np.linalg.matrix_rank()` 

Using this function, we can determine the rank of the matrix: 

\begin{bmatrix}
-2 & -5 & 8 & 0 & -17 \\
1 & 3 & -5 & 1 & 5 \\
3 & 11 & -19 & 7 & 1 \\
1 & 7 & -13 & 5 & -3 \\
\end{bmatrix}

```python
import numpy as np

# Define the matrix
matrix = [
    [-2, -5, 8, 0, -17],
    [1, 3, -5, 1, 5],
    [3, 11, -19, 7, 1],
    [1, 7, -13, 5, -3]
]

# Compute the rank of the matrix
rank = np.linalg.matrix_rank(matrix)

# Print the rank
print("Rank of the matrix:", rank)
```
